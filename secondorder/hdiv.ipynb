The function space $H(\operatorname{div})$
===
$\DeclareMathOperator{\opdiv}{div}$
We will
* give a clear definition of the space $H(\opdiv)$
* understand its boundary values, i.e. its trace
* understand its continuity properties

As for the Sobolev space $H^1$, we need a weak definition of the differential operator $\opdiv$. We call
$$
d = \opdiv \sigma
$$
the weak divergence iff
$$
\int_\Omega d \, \varphi = - \int_\Omega \sigma \nabla \varphi
$$
holds for all smooth testfunctions with compact support $\varphi \in C_0^\infty(\Omega)$.

We define the function space

$$
H(\opdiv) = \{ \sigma \in [L_2]^2 : \opdiv \sigma \in L_2 \}
$$

equipped with the norm

$$
\| \sigma \|_{H(\opdiv)} = \left( \| \sigma \|_{L_2}^2 + \| \opdiv \sigma \|_{L_2}^2 \right)^{1/2}.
$$

Smooth functions are dense in $H(\opdiv)$, i.e. $\forall \, \sigma \in H(\opdiv)$ and $\forall \varepsilon > 0$ there exists $\sigma^\varepsilon \in C^\infty$ such that

$$
\| \sigma - \sigma^\varepsilon \|_{H(\opdiv)} < \varepsilon
$$
(proven by commuting mollifiers).

For smooth functions $\sigma$ and $\varphi$ there holds the integration by parts formula

$$
\int_\Omega \opdiv \sigma \varphi = -\int_\Omega \sigma \nabla \varphi + \int_{\partial \Omega} \sigma_n \varphi
$$

Normal-trace of functions in $H(\opdiv)$
---

From the integration by parts formula applied to smooth functions $\sigma$ we get

\begin{eqnarray*}
\left| \int_{\partial \Omega} \sigma_n \varphi \right| 
& = & \left| \int_\Omega \opdiv \sigma \, \varphi + \int_\Omega \sigma \nabla \varphi \right| \\
& \leq & \| \opdiv \sigma \|_{L_2} \| \varphi \|_{L_2} + \| \sigma \|_{L_2} \| \nabla v \|_{L_2} \\
& \leq & \| \sigma \|_{H(\opdiv)} \| \varphi \|_{H^1}
\end{eqnarray*}

and thus

\begin{eqnarray*}
\| \sigma_n \|_{H^{-1/2}} & = &
\sup_{v \in H^{1/2}} \frac{\left< v, \sigma_n \right>}{\| v \|_{H^{1/2}}} \\
& \approx & 
\sup_{\varphi \in H^1(\Omega)} 
\frac{\left< \varphi_{|\partial \Omega}, \sigma_n 
\right>}{\|  \varphi \|_{H^1(\Omega)}} 
\leq \| \sigma \|_{H(\opdiv)}
\end{eqnarray*}

We have shown that

$$
\| \sigma_n \|_{H^{-1/2}} \prec \| \sigma \|_{H(\opdiv)} \quad \text{ for smooth $\sigma$ }
$$

Now we use density of smooth functions in $H(\opdiv)$ to uniquely extend the trace operator to the whole space $H(\opdiv)$.

Having well-defined boundary values, we get now the integration by parts formula for arbitrary functions $\sigma \in H(\opdiv)$.

$H(\opdiv)$ on sub-domains
---

Assume that we have a non-overlapping domain decomposition

$$
\Omega = \Omega_1 \cup \Omega_2 \ldots \Omega_n
$$

with interfaces

$$
\gamma_{ij} = \overline{\Omega_i} \cap \overline{\Omega_j}
$$

*Theorem:* for $\sigma \in [L_2(\Omega)]^n$ such that
* $\sigma_{|\Omega_i} \in H(\opdiv, \Omega_i)$ with $d_i = \opdiv_{\Omega_i} \sigma_{|\Omega_i}$
* $\sigma_{\Omega_i} n_i = - \sigma_{\Omega_j} n_j$ on $\gamma_{ij}$

then there is $\sigma \in H(\opdiv, \Omega)$ with

$$
(\opdiv \sigma)_{|\Omega_i} = d_i
$$

Proof: We apply the integration by parts formula on sub-domains, and observe that boundary terms cancel out:

\begin{eqnarray*}
\int_\Omega d \varphi & = & \sum_{\Omega_i} \int_{\Omega_i} d_i \varphi 
= \sum_{\Omega_i} \int_{\Omega_i} \opdiv_{\Omega_i} \sigma_{|\Omega_i} \varphi \\
& = & \sum_{\Omega_i} -\int_{\Omega_i} \sigma_{|\Omega_i} \nabla \varphi + \int_{\partial \Omega_i} n_i \sigma_{|\Omega_i} \varphi \\
& = & \int_\Omega \sigma \nabla \varphi
\end{eqnarray*}
